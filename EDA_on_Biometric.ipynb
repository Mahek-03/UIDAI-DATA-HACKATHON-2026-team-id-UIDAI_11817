# Merging all three datasets

```python
bio = pd.concat([bio_1, bio_2, bio_3], ignore_index=True)
bio
```

# State Contribution Analysis (Top States)

```python
state_auth = bio.groupby('state')[['bio_age_5_17','bio_age_17_']].sum()
state_auth['total'] = state_auth.sum(axis=1)

top_states = state_auth.sort_values('total', ascending=False).head(10)

plt.figure(figsize=(12,6))
top_states['total'].plot(kind='bar', color='Turquoise')
plt.title('Top 10 States by Aadhaar Biometric Authentications')
plt.xlabel('State')
plt.ylabel('Total Authentications')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

# Age Group Dominance Analysis (Stacked Bars)

```python
top_states = bio.groupby('state')[['bio_age_5_17','bio_age_17_']].sum().head(10)

top_states.plot(
    kind='bar',
    figsize=(12,6)
)

plt.title('Biometric Authentications by Age Group (Top States)')
plt.xlabel('State')
plt.ylabel('Total Authentications')
plt.xticks(rotation=45)
plt.legend(['Age 5–17', 'Age 17+'])
plt.show()
```

# Date-Wise Aadhaar Biometric Authentication Patterns

```python
import matplotlib.pyplot as plt
import pandas as pd

bio['date'] = pd.to_datetime(bio['date'])
daily = bio.groupby('date')[['bio_age_5_17','bio_age_17_']].sum()

plt.figure(figsize=(12,6))
plt.stackplot(
    daily.index,
    daily['bio_age_5_17'],
    daily['bio_age_17_'],
    labels=['Age 5–17', 'Age 17+'],
    colors=['#66c2a5', '#fc8d62']
)
plt.title('Daily Aadhaar Biometric Authentication Trend')
plt.xlabel('Date')
plt.ylabel('Total Authentications')
plt.legend(loc='upper left')
plt.show()
```

# District-Level Usage Intensity (Heatmap Analysis)

```python
import seaborn as sns
import matplotlib.pyplot as plt

district_usage = (
    bio.groupby('district')['bio_age_17_']
    .sum()
    .sort_values(ascending=False)
    .head(20)
)

plt.figure(figsize=(8,10))
sns.heatmap(
    district_usage.to_frame(),
    cmap='Reds',
    annot=True,
    fmt='.0f',
    linewidths=0.5
)

plt.title('District-Level Aadhaar Biometric Usage Intensity (Age 17+)')
plt.xlabel('Authentication Count')
plt.ylabel('District')
plt.show()
```

# Districtwise Contribution Analysis (Overall top 10)

```python
district_auth = bio.groupby('district')[['bio_age_5_17','bio_age_17_']].sum()
district_auth['total'] = district_auth.sum(axis=1)

top_districts = district_auth.sort_values('total', ascending=False).head(10)

plt.figure(figsize=(12,6))
plt.barh(top_districts.index, top_districts['total'])
plt.title('Top 10 Districts by Aadhaar Biometric Authentications')
plt.xlabel('Total Authentications')
plt.ylabel('District')
plt.show()
```
