### Create Total enrolments

```python
enrol['total_enrolment'] = (
    enrol['age_0_5'] +
    enrol['age_5_17'] +
    enrol['age_18_greater']
)
```

### Filter for Maharashtra → Mumbai & Pune

```python
city_df = enrol[
    (enrol['state'] == 'Maharashtra') &
    ((enrol['district'] == 'Mumbai') | (enrol['district'] == 'Pune'))
]
```

```python
city_df['district'].value_counts()
```

```python
city_summary = city_df.groupby('district')['total_enrolment'].sum()
city_summary
```

### STEP 5: Visualization


---



**State-wise Enrolment Comparison**

```python
state_summary = (
    enrol
    .groupby('state')['total_enrolment']
    .sum()
    .sort_values(ascending=False)
)
top_states = state_summary.head(10)

plt.figure(figsize=(8,5))
top_states.plot(kind='bar',color='#e84393')

plt.title('Top 10 States by Aadhaar Enrolment (in Millions)')
plt.xlabel('State')
plt.ylabel('Enrolments (Millions)')
plt.xticks(rotation=45, ha='right')
plt.grid(axis='y')
plt.tight_layout()
plt.show()
```

**Age-wise Aadhar Enrollment**

```python
age_summary = city_df.groupby('district')[['age_0_5','age_5_17','age_18_greater']].sum()
age_summary
```

```python
age_summary.plot(kind='bar', figsize=(7,4))
plt.title('Age-wise Aadhaar Enrolment by City')
plt.ylabel('Count')
plt.show()
```

**Monthly Aadhaar Enrolment Trend (Mumbai vs Pune)**

```python
print(city_df.columns)
city_df = city_df.copy()

city_df['date'] = pd.to_datetime(
    city_df['date'],
    errors='coerce' #converts bad values to NaT
)
```

```python
city_df['date'].dtype
```

```python
city_df = city_df.dropna(subset=['date'])
```

```python
city_df['month'] = city_df['date'].dt.to_period('M')
```

```python
monthly_trend = (
    city_df
    .groupby(['month','district'])['total_enrolment']
    .sum()
    .reset_index()
)

monthly_trend['month'] = monthly_trend['month'].astype(str)
```

```python
plt.figure(figsize=(10,5))
sns.lineplot(
    data=monthly_trend,
    x='month',
    y='total_enrolment',
    hue='district',
    marker='o',
    linewidth=3
)

plt.title('Monthly Aadhaar Enrolment Trend (Mumbai vs Pune)')
plt.xlabel('Month')
plt.ylabel('Total Enrolments')
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()
plt.show()
```

**Time vs City Enrolment HeatMap**

```python
heat_data = monthly.pivot(
    index='month',
    columns='district',
    values='total_enrolment'
)

plt.figure(figsize=(6,5))
sns.heatmap(
    heat_data,
    annot=True,
    fmt='.0f',
    cmap='YlOrRd'
)

plt.title('Enrolment Intensity Heatmap')
plt.ylabel('Month')
plt.xlabel('City')
plt.show()
```

**Top Pincode Analysis**

```python
top_pincode = (
    city_df
    .groupby(['district','pincode'])['total_enrolment']
    .sum()
    .reset_index()
    .sort_values('total_enrolment', ascending=False)
    .head(10)
)

plt.figure(figsize=(8,4))
sns.barplot(
    data=top_pincode,
    x='pincode',
    y='total_enrolment',
    hue='district'
)

plt.title('Top 10 Pincodes by Aadhaar Enrolment')
plt.xticks(rotation=45)
plt.show()
```

**Top 5 Cities Share in Aadhhar Enrolment(Maharshtra)**

```python
state_df = enrol[enrol['state'] == 'Maharashtra']
city_total = (
    state_df
    .groupby('district')['total_enrolment']
    .sum()
    .sort_values(ascending=False)
)

top5 = city_total.head(5)

plt.figure(figsize=(5,5))
plt.pie(
    top5,
    labels=top5.index,
    autopct='%1.1f%%',
    startangle=90,
    wedgeprops=dict(width=0.4)
)

plt.title(' Top 5 Cities Aadhaar Enrolment Share(Maharashtra)')
plt.show()
```

###📌 Remarks / Insights

🔹Aadhaar enrolment shows strong regional concentration, with a few states and major urban cities contributing a large share of total enrolments, indicating the influence of population density and urban accessibility.


🔹Adult enrolments (18+ years) dominate across regions, while younger age groups contribute relatively fewer enrolments, reflecting Aadhaar’s primary use in adult-oriented services and verifications.


🔹Monthly trend analysis (after date conversion and aggregation) reduced daily noise and revealed clearer patterns. Mumbai consistently records higher enrolment volumes than Pune, indicating sustained enrolment demand in larger metropolitan areas.


🔹Aggregating age-group data to create a total enrolment metric enabled meaningful comparison across states, cities, and pincodes, helping identify enrolment intensity, hotspots, and temporal trends.
